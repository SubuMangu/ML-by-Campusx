### **The fillna() method**
---

- fillna() is used to fill missing values

```python
import pandas as pd
data=pd.read_csv('matches - matches.csv')
```

```python
data.fillna(0).head()
```

- we can use fillna() on a particular columns

```python
#Showing error in the below code
#data[['umpire1','umpire2','umpire3']].fillna('Not Specified',inplace=True)
data[['umpire1','umpire2','umpire3']].fillna('Not Specified')[1:6]
```

```python
data[['umpire1','umpire2']].fillna(method='ffill')[1:6]
```

```python
data[['umpire1','umpire2']].fillna(method='bfill')[1:6]
```

```python
data[['umpire1','umpire2']].fillna(method='backfill')[1:6]
```
