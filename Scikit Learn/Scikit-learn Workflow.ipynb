**Value Imputation:**

- Deleteing features are never prefered since by eliminating a column, we are losing one factor that produces/contributes the outcome/output.This will eventually lead to genaralization error. 

- To solve this issue we use <span style="color:SkyBlue">SimpleImputer</span>

```python
from sklearn.impute import SimpleImputer
```

```python
print(f'Number of null values before imputing in age:{df.age.isnull().sum()}')
```

```python
imp=SimpleImputer(strategy='mean')
df['age']=imp.fit_transform(df[['age']]).reshape(-1,1)
```

```python
print(f'Number of null values in age column: {df.age.isnull().sum()}')
```

```python
def get_parameters(df):
    parameters={}
    for col in df.columns[df.isnull().any()]:
        if df[col].dtype=='float64' or df[col].dtype=='int64' or df[col].dtype=='int32':
            strategy='mean'
        else:
            strategy='most_frequent'
        missing_values=df[col][df[col].isnull()].values[0]
        parameters[col]={'missing_values':missing_values,'strategy':strategy}
    return parameters
```

```python
parameters=get_parameters(df)
```

```python
for col,param in parameters.items():
    x=param['missing_values']
    y=param['strategy']
    imp=SimpleImputer(missing_values=x,strategy=y)
    df[col]=imp.fit_transform(df[[col]])
```

**<u>Feature Engineering</u>:**

- 'sibsp' captures the number of siblings passengers have travelled with

```python
from turtle import title


df['family']=df['sibsp']+df['parch']
df.loc[df['family']>0,'travelled_alone']=0
df.loc[df['family']==0,'travelled_alone']=1
df['travelled_alone'].value_counts().plot(title='Passenger travelled alone',kind='bar')
```

**Data Scaling:**

- If the data in any conditions has data points far from each other, scaling is a technique to make them closer to each other.

- To normalise the data we can use 2 methods that are avilable in scikit-learn:
     1. <u>StandardScaler</u>: Standard Scalar follows the Standard Normal Distribution i.e., it assumes a normal distribution for data within each feature.
     2. <u>MinMaxScaler</u>:For each value in a feature, MinMaxScalarr subtracts the minimum value in the feature and then divides by the range.
     

```python
from sklearn.preprocessing import StandardScaler
```

```python
num_cols=df.select_dtypes(include=['int64','float64','int32']).columns
print(num_cols)
#Prints numerical columns
```

```python
ss=StandardScaler()
```

```python
df[num_cols]=ss.fit_transform(df[num_cols])
df[num_cols].describe()
#all the values are standardized
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
df[num_cols]=minmax.fit_transform(df[num_cols])
```
